```python
#Missing Values
df.isnull().sum()
```

```python
#Impute the missing values in review rating  with the median per each category since there are several categories e.g clothing,...etc
df["Review Rating"]=df.groupby("Category")["Review Rating"].transform(lambda x:x.fillna(x.median()))
 
```

```python
df.isnull().sum()
```

```python
#Ensuring consistencies
df.columns=df.columns.str.lower().str.replace(' ','_')
```

```python
df.head()
```

```python
#Renaming column names
df=df.rename(columns={"purchase_amount_(usd)":"purchase_amount"})
```

```python
df.head()
```

```python
#Labelling variables ,qcut splits the age grups into 4 equal groups
labels=["Young Adult","Adult","Middle-aged","Senior"]
df["age_group"]=pd.qcut(df["age"],q=4,labels=labels)
```

```python
df.head()
```

```python
#Checking two columns ata ago
df[["age","age_group"]].head(10)
```

```python
#Create a new column Purchase frequency days
frequency_mapping={"Fortnightly":14,"Weekly":7,"Monthly":30,"Quarterly":90,"Bi-weekly":14,"Annually":365,"Every 3 months":90}
df["purchase_frequency_days"]=df["frequency_of_purchases"].map(frequency_mapping)
```

```python
#Checking if two columns have the same values
(df["discount_applied"]==df["promo_code_used"]).all()
```

```python
#Drop the promo code column
df.drop("promo_code_used",axis=1)
```
